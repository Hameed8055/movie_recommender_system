# Converting Columns(genres+keywors+cast+crew) in understandable form where in out there should be no space between 2 words like James Cameron should be written as JamesCameron

```python
import ast

# Generic function for genres & keywords
def convert_to_list(text):
    try:
        data = ast.literal_eval(text)
        return [d['name'].replace(" ", "") for d in data]   # remove spaces inside names
    except:
        return []

# Function for top 3 cast
def get_top_cast(text):
    try:
        data = ast.literal_eval(text)
        return [d['name'].replace(" ", "") for d in data[:3]]
    except:
        return []

# Function for crew (only Director)
def get_director(text):
    try:
        data = ast.literal_eval(text)
        for d in data:
            if d.get('job') == 'Director':
                return [d['name'].replace(" ", "")]
        return []
    except:
        return []

# Apply transformations
final_df['genres'] = final_df['genres'].apply(convert_to_list)
final_df['keywords'] = final_df['keywords'].apply(convert_to_list)
final_df['cast'] = final_df['cast'].apply(get_top_cast)
final_df['crew'] = final_df['crew'].apply(get_director)

# Check results
print(final_df[['title', 'genres', 'keywords', 'cast', 'crew']].head())
```

# Merging overview+genres+keywords+cast+crew

```python
# Merge genres + keywords + cast + crew into one column "tags"
final_df['tags'] = final_df['genres'] + final_df['keywords'] + final_df['cast'] + final_df['crew']

# Convert list to string (no spaces between words, everything continuous)
final_df['tags'] = final_df['tags'].apply(lambda x: " ".join(x))
# Converting tags column into lower case
final_df["tags"]=final_df["tags"].apply(lambda x:x.lower())


# Keep only required columns
final_df = final_df[['id', 'title', 'tags']]

# Show result
print(final_df.head())
```

# Vectorize Text

>>🔎 Why We Need Vectorization

Computers can’t understand raw text like "Action Adventure Fantasy SciFi SamWorthington JamesCameron".

Machine Learning models and similarity calculations only work with numbers.

So we convert text → numerical form (vectors).

```python
from sklearn.feature_extraction.text import CountVectorizer

# Initialize CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')

# Fit and transform the tags column
vectors = cv.fit_transform(final_df['tags']).toarray()

print("Shape of vectors:", vectors.shape)
print("First row vector:", vectors[0])
```

# Checking cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute cosine similarity matrix
similarity = cosine_similarity(vectors)

print("Shape of similarity matrix:", similarity.shape)
print("Example similarity row for first movie:\n", similarity[0])
```

# Recommender Function

```python
def recommend(movie):
    # find index of the movie
    idx = final_df[final_df['title'] == movie].index[0]
    distances = similarity[idx]
    # sort movies by similarity (skip the first one since it’s the same movie)
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    for i in movie_list:
        print(final_df.iloc[i[0]].title)
        
# Example usage
recommend('Avatar')
```

```python
import pandas as pd
import numpy as np
import pickle
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# ========== STEP 1: Vectorization ==========
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(final_df['tags']).toarray()

# ========== STEP 2: Cosine Similarity ==========
similarity = cosine_similarity(vectors)

# ========== STEP 3: Save Data ==========
pickle.dump(final_df, open("movies.pkl", "wb"))
pickle.dump(similarity, open("similarity.pkl", "wb"))

print("✅ Files saved: movies.pkl & similarity.pkl")

# ========== STEP 4: Load Data ==========
movies = pickle.load(open("movies.pkl", "rb"))
similarity = pickle.load(open("similarity.pkl", "rb"))

# ========== STEP 5: Recommendation Function ==========
def recommend(movie):
    if movie not in movies['title'].values:
        print("❌ Movie not found in dataset")
        return
    idx = movies[movies['title'] == movie].index[0]
    distances = similarity[idx]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    print(f"\n🎬 Recommended movies for: {movie}\n")
    for i in movie_list:
        print(movies.iloc[i[0]].title)

# Example usage
recommend("Avatar")
```
